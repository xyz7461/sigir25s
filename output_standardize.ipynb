```python
import pandas as pd
```

```python
import os
import glob
import ast


directory = 'good'
my_files = [f for f in glob.glob(os.path.join(directory, "*.csv"))]
# infile = 'good/llama70b_rps.csv'

for infile in my_files:
    df = pd.read_csv(infile)
    sml = df.drop(['id', 'prompt', 'samples'], axis=1)
    sml.head()
    import ast
    def parse_json(row):
        parsed_json = ast.literal_eval(row)
        title = parsed_json.get('MovieName', '')
        predicted_rating = parsed_json.get('RatingPrediction', '')
        return title, predicted_rating
    sml[['title', 'predicted_rating']] = sml['json'].apply(parse_json).apply(pd.Series)
    sml['title'] = sml['title'].apply(lambda x: x.split(')-')[0]+')')
    fin = sml.drop(columns=['json'])
    outfile = infile.split('/')[-1]
    fin.to_csv('llm_output_processed/'+outfile, index=False)
```
